```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
us_open = pd.read_csv("2021-usopen-points.csv")
wimbledon = pd.read_csv("2021-wimbledon-points.csv")

# 提取所需列
us_open_selected = us_open[['P1DistanceRun', 'P2DistanceRun', 'RallyCount']]
wimbledon_selected = wimbledon[['P1DistanceRun', 'P2DistanceRun', 'RallyCount']]

# 计算 DistanceRun 和 RallyCount 的平均值
us_open_distance_avg = (us_open_selected['P1DistanceRun'] + us_open_selected['P2DistanceRun']).mean()
us_open_rally_avg = us_open_selected['RallyCount'].mean()

wimbledon_distance_avg = (wimbledon_selected['P1DistanceRun'] + wimbledon_selected['P2DistanceRun']).mean()
wimbledon_rally_avg = wimbledon_selected['RallyCount'].mean()

# 准备数据用于绘图
labels = ['DistanceRun', 'RallyCount']
us_open_values = [us_open_distance_avg, us_open_rally_avg]
wimbledon_values = [wimbledon_distance_avg, wimbledon_rally_avg]

x = range(len(labels))  # x轴位置

# 绘制对比柱状图
plt.figure(figsize=(8, 6))
plt.bar(x, us_open_values, width=0.4, label='US Open', color='#D54A3B', align='center')
plt.bar([p + 0.4 for p in x], wimbledon_values, width=0.4, label='Wimbledon', color='#579AC3', align='center')

# 设置图例和标签
plt.xlabel("Metrics", fontsize=12)
plt.ylabel("Average Values", fontsize=12)
plt.title("Comparison of Average Distance Run and Rally Count", fontsize=14)
plt.xticks([p + 0.2 for p in x], labels, fontsize=12)
plt.legend()
plt.tight_layout()

# 显示图像
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
us_open = pd.read_csv("2021-usopen-points.csv")
wimbledon = pd.read_csv("2021-wimbledon-points.csv")

# 提取所需列
us_open_selected = us_open[['P1DistanceRun', 'P2DistanceRun', 'RallyCount']]
wimbledon_selected = wimbledon[['P1DistanceRun', 'P2DistanceRun', 'RallyCount']]

# 计算 DistanceRun 和 RallyCount 的平均值
us_open_distance_avg = (us_open_selected['P1DistanceRun'] + us_open_selected['P2DistanceRun']).mean()
us_open_rally_avg = us_open_selected['RallyCount'].mean()

wimbledon_distance_avg = (wimbledon_selected['P1DistanceRun'] + wimbledon_selected['P2DistanceRun']).mean()
wimbledon_rally_avg = wimbledon_selected['RallyCount'].mean()

# 准备数据用于绘图
labels = ['US Open', 'Wimbledon']
distance_values = [us_open_distance_avg, wimbledon_distance_avg]
rally_values = [us_open_rally_avg, wimbledon_rally_avg]

# 创建子图
fig, axs = plt.subplots(1, 2, figsize=(8, 8), dpi=2000)

width = 0.8
x = range(len(labels))  # x轴位置
# x = []

# 绘制 DistanceRun 对比柱状图，指定标签
axs[0].bar(x, distance_values, color=['#D54A3B', '#579AC3'], width=width, label='Average Distance Run')
axs[0].set_xticks(x)
axs[0].set_xticklabels(labels)
axs[0].set_xlabel("Tournaments", fontsize=12)
axs[0].set_ylabel("Average Distance Run", fontsize=12)
axs[0].set_ylim(0, 55)
axs[0].set_title("Comparison of Average Distance Run", fontsize=14)
axs[0].grid(visible=True, linestyle='--', alpha=0.3)
axs[0].legend()

# 绘制 RallyCount 对比柱状图，指定标签
axs[1].bar(x, rally_values, color=['#D54A3B', '#579AC3'], width=width, label='Average Rally Count')
axs[1].set_xticks(x)
axs[1].set_xticklabels(labels)
axs[1].set_xlabel("Tournaments", fontsize=12)
axs[1].set_ylabel("Average Rally Count", fontsize=12)
axs[1].set_ylim(0, 3.7)
axs[1].set_title("Comparison of Average Rally Count", fontsize=14)
axs[1].grid(visible=True, linestyle='--', alpha=0.3)
axs[1].legend()

# 调整布局
plt.tight_layout()

# 显示图像
plt.show()
```
